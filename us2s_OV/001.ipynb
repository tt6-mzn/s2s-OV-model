```python
import random
import numpy as np
import matplotlib.pyplot as plt
import us2s_OV_CA
```

```python
def plot_pattern(model, ni, nf):
	x = np.array([[j for i in range(model.K)] for j in range(ni, nf+1)]) \
		.flatten()
	y = model.x[ni:nf+1, :model.K].flatten()
	plt.figure(figsize=(6.4, 6.4))
	plt.title(
		"L={}, K={}, n0={}, x0={}, v0={}, dt={}"
			.format(model.L, model.K, model.n_0, model.x_0, model.v_0, model.dt)
	)
	plt.xlabel("Time")
	plt.ylabel("Location of Vehicles")
	plt.scatter(x, y, s=1, c="black")
	plt.show()
```

```python
model = us2s_OV_CA.us2s_OV_CA(
	L=100,
	K=30,
	n_0=3,
	v_0=3,
	x_init=sorted(random.sample([i for i in range(100)], 30)),
	n_max=1010,
	x_0=1,
	dt=1
)
```

```python
model.simulate()
```

```python
plot_pattern(model, 0, 100)
```
