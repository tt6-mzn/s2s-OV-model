```python
model = us2s_OV_CA.us2s_OV_CA(
	L=100,
	K=30,
	n_0=3,
	v_0=3,
	x_init=sorted(random.sample([i for i in range(100)], 30)),
	n_max=1010,
	x_0=1,
	dt=1
)
```

```python
model.simulate()
```

```python
plot_pattern(model, 0, 100, line=True)
```
