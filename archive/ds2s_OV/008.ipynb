```python
import random
import numpy as np
import matplotlib.pyplot as plt
from ds2s_OV_x0 import ds2s_OV_x0 as ds2s_OV
from funcs import *
```

```python
model = ds2s_OV(
	L=10.0,
	K=25,
	n_0=5,
	x_0=0.1,
	v_0=0.7,
	dt=1.0,
	dx=1.0,
	x_init=np.array([4.0 + 0.1 * i for i in range(25)]),
	n_max=10000,
)
model.simulate()
plot_pattern(model, 0, 100)
```

```python
density = []
flow = []
for K in range(1, 101):
	# 渋滞を一つ作る
	x_init = np.array([0.1 * i for i in range(K)])
	model = ds2s_OV(
		L=10.0,
		K=K,
		n_0=5,
		x_0=0.1,
		v_0=0.7,
		dt=1.0,
		dx=0.1,
		x_init=x_init,
		n_max=2010,
	)
	model.simulate()
	density.append(model.density())
	flow.append(model.flow(1900, 2000))
	# 渋滞を二つ作る
	x_init = np.array([0.1 * i for i in range(K // 2)] + [0.1 * i + 5.0 for i in range((K + 1) // 2)])
	model = ds2s_OV(
		L=10.0,
		K=K,
		n_0=5,
		x_0=0.1,
		v_0=0.7,
		dt=1.0,
		dx=0.1,
		x_init=x_init,
		n_max=2010,
	)
	model.simulate()
	density.append(model.density())
	flow.append(model.flow(1900, 2000))
	# 等間隔に並べる
	x_init = np.array([10.0 / K * i for i in range(K)])
	model = ds2s_OV(
		L=10.0,
		K=K,
		n_0=5,
		x_0=0.1,
		v_0=0.7,
		dt=1.0,
		dx=0.1,
		x_init=x_init,
		n_max=2010,
	)
	model.simulate()
	density.append(model.density())
	flow.append(model.flow(1900, 2000))
	# ランダムに並べる
	for _ in range(10):
		x_init = np.array(random.sample([0.1 * i for i in range(100)], K))
		model = ds2s_OV(
			L=10.0,
			K=K,
			n_0=5,
			x_0=0.1,
			v_0=0.7,
			dt=1.0,
			dx=0.1,
			x_init=x_init,
			n_max=2010,
		)
		model.simulate()
		density.append(model.density())
		flow.append(model.flow(1900, 2000))
```

```python
plot_fundamental(density, flow)
```
