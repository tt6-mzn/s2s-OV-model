```python
import random
import numpy as np
import matplotlib.pyplot as plt
from ds2s_OV_revised import ds2s_OV_revised as ds2s_OV
from ds2s_OV_sort import ds2s_OV_sort
from ds2s_OV_x0 import ds2s_OV_x0
```

```python
def plot_pattern(model, ni, nf, line=False):
    x = np.array([[j for i in range(1 if line else 0, model.K)] for j in range(ni, nf+1)]) \
        .flatten()
    y = model.x[ni:nf+1, 1 if line else 0:model.K].flatten()
    plt.figure(figsize=(6.4, 6.4))
    plt.title(
        "L={}, K={}, n0={}, x0={}, v0={}, dt={}, dx={}"
            .format(model.L, model.K, model.n_0, model.x_0, model.v_0, model.dt, model.dx)
    )
    plt.xlabel("Time")
    plt.ylabel("Location of Vehicles")
    plt.scatter(x, y, s=1, c="black")
    if line:
        x = np.array([j for j in range(ni, nf+1)])
        y = model.x[ni:nf+1, 0]
        plt.plot(x, y, color="black")
    plt.show()
```

```python
model = ds2s_OV(
	L=10.0,
	K=25,
	n_0=5,
	x_0=0.1,
	v_0=0.7,
	dt=1.0,
	dx=1.0,
	x_init=np.array([4.0 + 0.1 * i for i in range(25)]),
	n_max=10000,
)
model.simulate()
plot_pattern(model, 930, 1030, line=True)
```

```python
model = ds2s_OV_sort(
	L=10.0,
	K=25,
	n_0=5,
	x_0=0.1,
	v_0=0.7,
	dt=1.0,
	dx=1.0,
	x_init=np.array([4.0 + 0.1 * i for i in range(25)]),
	n_max=10000,
)
model.simulate()
plot_pattern(model, 930, 1030)
```

```python
model = ds2s_OV_x0(
	L=10.0,
	K=60,
	n_0=5,
	x_0=0.1,
	v_0=0.7,
	dt=1.0,
	dx=1.0,
	x_init=np.array([0.1 * i for i in range(30)] + [5.0 + 0.1 * i for i in range(30)]),
	n_max=1200,
)
model.simulate()
plot_pattern(model, 0, 100)
```
