```python
x_min = 0.1  # 車両の長さ: 5.0mで換算

L = 10.0  # レーンの長さ: 470m
n_0 = 5

x_0 = 0.15  # 最小車間距離: 車両の長さ+3mで換算
v_0 = 0.5  # 車両の最高速度: 法定速度100km/hで換算
dt = 1.0  # 時間差分1秒
dx = 0.1  # テキトー。基本図がいい感じになるように調整
```
